## The above image shows two problem to deep learning 

### Vanishing Gradient :- The gradient is so small that model is not learning at all (weights do not change) and this lead to bad solution
### Exploding Gradient :- The gradient befor good minima is so sharpe(very high mangnitue) that the next step taken over-shoots the good minima
