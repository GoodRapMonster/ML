```python
import numpy as np 
from scipy.signal import convolve2d
from imageio import imread
import matplotlib.pyplot as plt

print("Success")
```

```python
# Manually making the image and applying the convolution
imgN=20
image=np.random.randn(imgN,imgN)
# print(image)

KernalN=7
Y,X=np.meshgrid(np.linspace(-3,3,KernalN),np.linspace(-3,3,KernalN))

kernal=np.exp(-(X**2+Y**2)/7)

fig,ax=plt.subplots(1,2,figsize=(5,4))

ax[0].imshow(image)
ax[1].imshow(kernal)

plt.show()
```

```python
convoutput=np.zeros((imgN,imgN))
halfkr=KernalN//2

for rowi in range(halfkr,imgN-halfkr):
    for coli in range(halfkr,imgN-halfkr):
        pieceOfImg=image[rowi-halfkr:rowi+halfkr+1,:]
        pieceOfImg=pieceOfImg[:,coli-halfkr:coli+halfkr+1]

        dotproduct=np.sum(pieceOfImg*kernal[::-1,::-1])

        convoutput[rowi,coli]=dotproduct
```

```python
convoutput2=convolve2d(image,kernal,mode='valid')
```

```python
fig,ax=plt.subplots(2,2,figsize=(8,8))

ax[0,0].imshow(image)
ax[0,1].imshow(kernal)
ax[1,0].imshow(convoutput)
ax[1,1].imshow(convoutput2)

plt.show()
```
