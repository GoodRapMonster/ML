````python
# lets define some Vector but we will use numpy as it prodive more funcationlity and written in C so its going to be fast

v1=np.array([1,2,3])# np.array always expect a list as input 
print(v1,type(v1))

print('''NOTE: vector V1 is dimensionless which means it can neither be a row vector nor column vector''')
print(v1.ndim)
print(f"{v1.shape}, This shows us it is dimensionless")
````

````python
#Vector with dimensions (row,column) are defined as :

v1_row=np.array([[1,2,3]])
print(f"{v1_row}<- Vector\n{v1_row.ndim}<-Dimensions it have \n{v1_row.shape}<-Shape of the vector")

v1_column=np.array([[1],[2],[3]])
print(f"\n\n{v1_column}<- Vector\n{v1_column.ndim}<-Dimensions it have \n{v1_column.shape}<-Shape of the vector")
````

## Operations on Vectors 

````python
#addition

v1=np.array([1,2,3]) # dimensionless vector 
v2=np.array([[1,2,3]]) # row Vector 
v3=np.array([[1],[2],[3]]) # column vector 

v4=np.array([[1,2]]) # row vector with different shape 
v5=np.array([[1],[2]])# column vector with different shape

print(f"{v1+v2}\n")
print(f"{v1+v3}\n")
print(f"{v2+v3}\n")
print(f"{v2+v4}\n")
````
